```python

import gender_guesser.detector as gender
import pandas as pd
import csv

d = gender.Detector(case_sensitive=False)

#print(d.get_gender(u"Ann"))

openOutputFile = open("outputwithgenderFile.csv", 'a')

data = pd.read_csv("Sample first names from new adds.csv")
df = pd.DataFrame(data)

df1 = df.dropna()
df1['First name'] = df1['First name'].str.replace(',', ' ').str.replace('.', ' ').str.replace('-', ' ').str.replace(' ', '')
# names.gender = " "
# df['First name'].dropna(inplace=True)
headers = ["Firstname", "Gender", "\n"]
csvWriter = csv.writer(openOutputFile, delimiter=',', lineterminator='')
csvWriter.writerow(headers)

for i in df1["First name"]:
    gendername = d.get_gender(i)
    print(i, gendername)
    csvWriter.writerow([i, gendername, "\n"])
    
openOutputFile.close()
    
#     newdata_gender = pd.DataFrame()
#     newdata_gender["First name"] = i
#     newdata_gender["gendername"] = gendername
#     df1["gendername"] = gendername
#     newdata_gender.to_csv('genderout_sample.csv', index=False)
```

```python
import random
import os
import re
os.system('cls' if os.name=='nt' else 'clear')
while (1 < 2):
    print ("\n")
    print ("Rock, Paper, Scissors - Shoot!")
    userChoice = input("Choose your weapon [R]ock], [P]aper, or [S]cissors:")
    if not re.match("[SsRrPp]", userChoice):
        print ("Please choose a letter:")
        print ("[R]ock, [S]cissors or [P]aper.")
        continue
#     // Echo the user's choice
    print ("You chose: " + userChoice)
    choices = ['R', 'P', 'S']
    opponenetChoice = random.choice(choices)
    print ("I chose: " + opponenetChoice)
    if opponenetChoice == str.upper(userChoice):
        print ("Tie! ")
    #if opponenetChoice == str("R") and str.upper(userChoice) == "P"
    elif opponenetChoice == 'R' and userChoice.upper() == 'S':      
        print ("Scissors beats rock, I win! ")
        continue
    elif opponenetChoice == 'S' and userChoice.upper() == 'P':      
        print ("Scissors beats paper! I win! ")
        continue
    elif opponenetChoice == 'P' and userChoice.upper() == 'R':      
        print ("Paper beat rock, I win! ")
        continue
    else:       
        print ("You win!")
```
